```python
import pandas as pd
```

```python
d = pd.read_csv("./2023/data/day-2-part1.csv", sep=":").assign(
    gameid=lambda x: x.gameid.str.replace("Game ", "").astype(int),
)
for color in ["red", "blue", "green"]:
    d[color] = d.record_of_games.str.findall(rf"(\d+ {color})").apply(
        lambda x: [int(item.replace(f" {color}", "")) for item in x]
    )
print("Full data\n")
display(d.head())
mask = (
    d.red.apply(lambda x: all([item <= 12 for item in x]))
    & d.blue.apply(lambda x: all([item <= 14 for item in x]))
    & d.green.apply(lambda x: all([item <= 13 for item in x]))
)
print("Masked data\n")
display(d[mask].head())
print(f"Sum of ID games: {d[mask].gameid.sum()}")
```
